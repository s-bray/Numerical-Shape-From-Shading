```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the images
image1 = cv2.imread('img1.jpeg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('img2.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image1,cmap='gray')
plt.show()
```

```python
image2 = cv2.flip(image2, 1)
plt.imshow(image2,cmap='gray')
plt.show()
```

```python
new_width = 128
new_height = 128
image1 = cv2.resize(image1, (new_width, new_height))
image2 = cv2.resize(image2, (new_width, new_height))
# Compute the absolute difference between the images
difference = np.abs(image1.astype(np.float32) - image2.astype(np.float32))

# Normalize the difference to the range [0, 1]
difference_normalized = difference / np.max(difference)

# Display the heatmap
plt.imshow(difference_normalized, cmap='coolwarm')
plt.colorbar()  # Add colorbar to show intensity scale
plt.title('Heatmap between Image 1 and Image 2')
plt.show()
```
